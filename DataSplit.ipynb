```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Directories containing annotations and images
annotations_dir = "/home/anil/Downloads/Helmet/Data/annotations_txt"
images_dir = "/home/anil/Downloads/Helmet/Data/images_png/"

# Define output directories for annotations
train_dir = "/home/anil/Downloads/Helmet/src/labels/train"
test_dir = "/home/anil/Downloads/Helmet/src/labels/test"
valid_dir = "/home/anil/Downloads/Helmet/src/labels/valid"

# Define output directories for images
train_dir_image = "/home/anil/Downloads/Helmet/src/images/train"
test_dir_image = "/home/anil/Downloads/Helmet/src/images/test"
valid_dir_image = "/home/anil/Downloads/Helmet/src/images/valid"

# Create output directories if they don't exist
for directory in [train_dir, test_dir, valid_dir, train_dir_image, test_dir_image, valid_dir_image]:
    os.makedirs(directory, exist_ok=True)

# List all annotations files
annotation_files = sorted(os.listdir(annotations_dir))
image_files = sorted(os.listdir(images_dir))

# Ensure that the number of annotation files matches the number of image files
assert len(annotation_files) == len(image_files), "Number of annotation files does not match number of image files"

# Split the data into train, test, and validation sets
train_files, test_valid_files = train_test_split(annotation_files, test_size=0.3, random_state=42)
test_files, valid_files = train_test_split(test_valid_files, test_size=0.5, random_state=42)

# Function to move files to destination directory
def move_files(file_list, src_dir, dest_dir):
    for file in file_list:
        src_file = os.path.join(src_dir, file)
        dest_file = os.path.join(dest_dir, file)
        if os.path.exists(src_file):
            shutil.move(src_file, dest_file)
        else:
            print(f"File '{file}' does not exist in the source directory.")

# Move annotation files to respective split directories
move_files(train_files, annotations_dir, train_dir)
move_files(test_files, annotations_dir, test_dir)
move_files(valid_files, annotations_dir, valid_dir)
```

```python
# list_img = []
# Loop through image files in the source directory
for filename in os.listdir(images_dir):
    if filename.endswith(".png"):
        # Remove the .png extension from the filename
        img_name = os.path.splitext(filename)[0]
        # list_img.append(img_name)
    
        if (img_name + '.txt') in os.listdir(train_dir):
            # print('true_train')
            src_file = os.path.join(images_dir, filename)
            dest_file = os.path.join(train_dir_image, filename)
            shutil.move(src_file, dest_file)
            
        elif (img_name + '.txt') in os.listdir(test_dir):
            # print('true_test')
            src_file = os.path.join(images_dir, filename)
            dest_file = os.path.join(test_dir_image, filename)
            shutil.move(src_file, dest_file)
            
        elif (img_name + '.txt') in os.listdir(valid_dir):
            # print('true_val')
            src_file = os.path.join(images_dir, filename)
            dest_file = os.path.join(valid_dir_image, filename)
            shutil.move(src_file, dest_file)
                        

```
